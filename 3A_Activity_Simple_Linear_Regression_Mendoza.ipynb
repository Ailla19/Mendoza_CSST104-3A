Simple Linear Regression

# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# 2. Generating Random Data:

```python
num_points = 50

x = np.random.randint(0, 100, num_points)
y = 2 + x + np.random.normal(0, 10, num_points)
```

# 3. Creating Data Frame

```python
data = pd.DataFrame({'x': x,
                    'y': y})
data.to_csv('data.csv', index = False)

print(data.head())
```

# 4. Visualizing the Data



```python
plt.scatter(data['x'], data['y'])
plt.xlabel('Dependent Variable')
plt.ylabel('Independent Variable')
plt.title('Scatter Plot of Data')
plt.show()
```

# 5. Preparing data for Modeling

```python
x = data[['x']] # Independent Variable
y = data[['y']] # Dependent Variable

```

# 6. Fitting the Linear Regression Model

```python
model = LinearRegression()

model.fit(x,y)
```

# 7. getting Model Coefficients



```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope',slope)
print('Interceot', intercept)
```

# 8. Visualizing the Linear Regression Line

```python
plt.scatter(x,y, color = 'pink')
plt.plot(x, model.predict(x), color = 'violet', marker = '+')
plt.xlabel('Dependent Variable')
plt.ylabel('Independent Variable')
plt.title('Linear Regression')
plt.show()
```

# 9. Making Prediction

```python
new_x = [[np.random.randint(100,200)]]

predicted_y = model.predict(new_x)

print(f'predicted Value of y: {predicted_y}')
print(f'predicted value of x: {new_x}')
```
