```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/Lesson_4_student_grades_dataset 2.csv')
```

```python
df. columns
```

```python
x = df['Travel_Time_to_School']
y = df['Attendance_Percentage']

plt.scatter(x, y, color = 'violet')
plt.xlabel('Independent Variables')
plt.ylabel('Dependent Variables')
plt.show()
```

```python
x = df[['Travel_Time_to_School']] # Independent Variable
y = df[['Attendance_Percentage']] # Dependent Variable

model = LinearRegression()

model.fit(x,y)
```

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope',slope)
print('Interceot', intercept)
```

```python
plt.scatter(x,y, color = 'pink')
plt.plot(x, model.predict(x), color = 'violet',)
plt.xlabel('Dependent Variable')
plt.ylabel('Independent Variable')
plt.title('Linear Regression')
plt.show()
```

```python
new_x = [[np.random.randint(100,200)]]

predicted_y = model.predict(new_x)

print(f'predicted Value of y: {predicted_y}')
print(f'predicted value of x: {new_x}')
```
