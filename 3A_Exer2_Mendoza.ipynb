1. Data Loading

```python
import pandas as pd

df = pd.read_csv('/content/sample_data/hardwareStore.csv')
```

2. Initial Data Analysis

```python
statistics = df.describe()

statistics

head = df.head(22)
```

3. Category Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sne

plt.figure(figsize = (10, 5))

plt.subplot(1, 2, 1)
sne.countplot(data = df, x = 'CATEGORY_NAME')
plt.title('Distribution of the Category Name')

plt.figure(figsize = (9, 5))

plt.subplot(1, 2, 1)
sne.histplot(data = df, x = 'CATEGORY_ID')
plt.title('Distribution of the Category Id')



```

4. Cost and Price Analysis

```python
average_standard_cost = df['STANDARD_COST'].mean()
average_list_price = df['LIST_PRICE'].mean()

highest_standard_cost = df['STANDARD_COST'].mean()
highest_list_price = df['LIST_PRICE'].mean()

print(average_standard_cost)
print(average_list_price )
```

5. Location Analysis

```python
state_stock = df.groupby(['STATE']).sum()
max_qauntity_state = state_stock['QUANTITY'].idxmax()

print("State with the highest quantity of products in stock", max_qauntity_state)
```
