 Data Loading and *Preprocessing*

```python
import pandas as pd
df = pd.read_csv('/content/onlinefoods.csv')

missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
print("First few rows of the dataset:")
print(df.head())
print("Descriptive Statistics for Numeric Columns:")
print(df.describe())
```

 Exploratory Data Analysis (EDA)


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Feedback', multiple='stack', kde=True)
plt.title('Distribution of Age and its impact on Feedback')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Feedback')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Monthly Income', hue='Feedback')
plt.title('Proportions of Feedback across different levels of Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Count')
plt.legend(title='Feedback')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
